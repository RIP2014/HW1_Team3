```python
import copy
```

```python
class LocationElement():
    def __init__(self, char):
        self.char = char
        self.location = ''
    
    def __repr__(self):
        return self.char
    
    def __str__(self):
        return self.char

class WorldElement():
    CLEAR_CELL   = LocationElement('_')
    BLOCKED_CELL = LocationElement('X')
    ROBOT_CELL   = LocationElement('r')
    BOX_CELL     = LocationElement('o')
    BOX_A_CELL   = LocationElement('a')
    BOX_B_CELL   = LocationElement('b')
    BOX_C_CELL   = LocationElement('c')
    GOAL_CELL    = LocationElement('g')
    GOAL_A_CELL  = LocationElement('A')
    GOAL_B_CELL  = LocationElement('B')
    GOAL_C_CELL  = LocationElement('C')
    
#     def __init__(self, char):
#         self.char = char
#         self.location = None
    
#     def __repr__(self):
#         return self.char
    
#     def __str__(self):
#         return self.char
    
world_dect =    {'_': WorldElement.CLEAR_CELL, 
                 'X': WorldElement.BLOCKED_CELL,
                 'r': WorldElement.ROBOT_CELL,
                 'o': WorldElement.BOX_CELL,
                 'a': WorldElement.BOX_A_CELL,
                 'b': WorldElement.BOX_B_CELL,
                 'c': WorldElement.BOX_C_CELL,
                 'g': WorldElement.GOAL_CELL,
                 'A': WorldElement.GOAL_A_CELL,
                 'B': WorldElement.GOAL_B_CELL,
                 'C': WorldElement.GOAL_C_CELL}

start_dect =    {'_': WorldElement.CLEAR_CELL, 
                 'X': WorldElement.BLOCKED_CELL,
                 'r': WorldElement.ROBOT_CELL,
                 'o': WorldElement.BOX_CELL,
                 'a': WorldElement.BOX_A_CELL,
                 'b': WorldElement.BOX_B_CELL,
                 'c': WorldElement.BOX_C_CELL,
                 'g': WorldElement.CLEAR_CELL,
                 'A': WorldElement.CLEAR_CELL,
                 'B': WorldElement.CLEAR_CELL,
                 'C': WorldElement.CLEAR_CELL}

goal_dect =     {'_': WorldElement.CLEAR_CELL, 
                 'X': WorldElement.BLOCKED_CELL,
                 'r': WorldElement.ROBOT_CELL,
                 'o': WorldElement.CLEAR_CELL,
                 'a': WorldElement.CLEAR_CELL,
                 'b': WorldElement.CLEAR_CELL,
                 'c': WorldElement.CLEAR_CELL,
                 'g': WorldElement.GOAL_CELL,
                 'A': WorldElement.BOX_A_CELL,
                 'B': WorldElement.BOX_B_CELL,
                 'C': WorldElement.BOX_C_CELL}

block_dect =    {'o': WorldElement.BOX_CELL,
                 'a': WorldElement.BOX_A_CELL,
                 'b': WorldElement.BOX_B_CELL,
                 'c': WorldElement.BOX_C_CELL,
                 'g': WorldElement.BOX_CELL}

class WorldDirections():
    UP     = ('up')
    DOWN   = ('down')
    LEFT   = ('left')
    RIGHT  = ('right')
    
    def __init__(self, char):
        self.char = char
    
    def __repr__(self):
        return self.char
    
    def __str__(self):
        return self.char
```

```python
class Sokoban(object):
    def __init__(self, world_path, labeled_boxes):
        """Initialize Sokoban game"""
        self.world_path = world_path
        self.labeled_boxes = labeled_boxes
        self.world       = self.loadWorld(world_path, labeled_boxes, map_type= 'original')
        self.start_world = self.loadWorld(world_path, labeled_boxes, map_type= 'start')
        self.goal_world  = self.loadWorld(world_path, labeled_boxes, map_type= 'goal')
        self.adjacency = [(i,j) for i in (-1,0,1) for j in (-1,0,1) if not (i == j == 0)] #the adjacency matrix
        
    def loadWorld(self, world_path, labeled_boxes, map_type):
        """Load the Sokoban world from ascii file"""
        with open(world_path) as f:
            staric_world = f.read()
        world = [list(row) for row in staric_world.split('\n')][:-1]
        for i, row in enumerate(world):
            for j, col in enumerate(row):
                if(map_type == 'original'):
                    loc = copy.deepcopy(world_dect[world[i][j]])
                if(map_type == 'start'):
                    loc = copy.deepcopy(start_dect[world[i][j]])
                if(map_type == 'goal'):
                    loc = copy.deepcopy(goal_dect[world[i][j]])
                loc.location = 'r' + str(i) + '-' + 'c' + str(j)
                world[i][j] = copy.deepcopy(loc)
                    
        return world
    
    def writeProblem(self, problem_path):
        with open(problem_path, 'w') as f:
            map_comment = ''
            for i, row in enumerate(self.world):
                row_str = map(str, row)
                row_str = ''.join(row_str)
                map_comment += '; ' + row_str + ' \n'
            
            proble_comment = '; This is an auto generated sokoban problem'
            problem = '(define (problem sokoban_problem )'
            domain_comment = '; The domain for our sokoban problem is simply named sokoban'
            domain  = '(:domain sokoban)'
            f.write(map_comment + '\n')
            f.write(proble_comment + '\n')
            f.write(problem + '\n\n')
            f.write(domain_comment + '\n')
            f.write(domain + '\n\n')

            #Specify list of objects
            objects_comment = '; Specify list of objects'
            objects = '(:objects '
            f.write(objects_comment + '\n')
            f.write(objects + '\n\n')
            
            #Specify list of directions
            directions_comment = '; Specify list of directions'
            directions = '' + \
                str(WorldDirections.UP)     + ' ' + \
                str(WorldDirections.DOWN)   + ' ' + \
                str(WorldDirections.LEFT)   + ' ' + \
                str(WorldDirections.RIGHT)
            f.write(directions_comment + '\n')
            f.write(directions + '\n\n')
            
            #Specify list of blocks
            blocks_comment = '; Specify list of blocks'
            blocks = ''
            for i, row in enumerate(self.start_world):
                for j, col in enumerate(row):
                    loc = self.start_world[i][j]
                    if(loc.char in block_dect):
                        blocks += loc.char + ' '
            f.write(blocks_comment + '\n')
            f.write(blocks + '\n\n')
            
            #Specify list of locations
            locations_comment = '; Specify list of locations'
            locations = ''
            for i, row in enumerate(self.start_world):
                for j, col in enumerate(row):
                    loc = self.start_world[i][j]
                    locations += loc.location
                    locations += ' '
                locations += '\n'
            
            f.write(locations_comment + '\n')
            f.write(locations + '\n')
            
            #End of list of objects
            objects = ')'
            f.write(objects + '\n\n\n')
            
            
            
            
            
            
            #Init objects
            Init_comment = '; Init direction objects'
            Init = '(:init'
            f.write(Init_comment + '\n')
            f.write(Init + '\n\n')
            
            
            #Init direction objects
            directions_comment = '; Init direction objects'
            directions = '' + \
                '(direction ' + str(WorldDirections.UP) + ')\n' + \
                '(direction ' + str(WorldDirections.DOWN) + ')\n' + \
                '(direction ' + str(WorldDirections.LEFT) + ')\n' + \
                '(direction ' + str(WorldDirections.RIGHT) + ')'
                
            f.write(directions_comment + '\n')
            f.write(directions + '\n\n')

            #Init block objects and locations
            blocks_comment = '; Init block objects and locations'
            blocks = ''
            for i, row in enumerate(self.start_world):
                for j, col in enumerate(row):
                    loc = self.start_world[i][j]
                    if(loc.char in block_dect):
                        blocks += '(block ' + loc.char + ')\n'
                        blocks += '(block-at ' + loc.char +' '+ loc.location + ')\n'
            f.write(blocks_comment + '\n')
            f.write(blocks + '\n\n')

            #Init robot location
            robot_comment = '; Init robot location'
            robot = ''
            for i, row in enumerate(self.start_world):
                for j, col in enumerate(row):
                    loc = self.start_world[i][j]
                    if(loc.char == str(WorldElement.ROBOT_CELL)):
                        robot += '(robot-at ' + str(loc.location) + ')\n'
            
            f.write(robot_comment + '\n')
            f.write(robot + '\n')
            

            #Init navigable objects
            navigable_comment = '; Init navigable objects'
            navigable = ''
            for i, row in enumerate(self.start_world):
                for j, col in enumerate(row):
                    loc = self.start_world[i][j]
                    if(loc.char != str(WorldElement.BLOCKED_CELL)):
                        navigable += '(navigable ' + str(loc.location) + ')\n'
            
            f.write(navigable_comment + '\n')
            f.write(navigable + '\n')

            #Init empty objects
            empty_comment = '; Init empty objects'
            empty = ''
            for i, row in enumerate(self.start_world):
                for j, col in enumerate(row):
                    loc = self.start_world[i][j]
                    if(loc.char == str(WorldElement.CLEAR_CELL)):
                        empty += '(empty ' + str(loc.location) + ')\n'
            
            f.write(empty_comment + '\n')
            f.write(empty + '\n')

            #Init adjacent objects
            adjacent_comment = '; Init adjacent objects'
            adjacent = ''
            for i, row in enumerate(self.start_world):
                for j, col in enumerate(row):
                    adjacent += self.get_adjacents(i,j)
            
            f.write(adjacent_comment + '\n')
            f.write(adjacent + '\n')
            
            #Init Done
            done_comment = '; Init Done'
            done = ')'
            f.write(done_comment + '\n')
            f.write(done + '\n')

            #Define goal states
            goal_comment = '; Define goal states'
            goal = '(:goal '
            goals = []
            for i, row in enumerate(self.goal_world):
                for j, col in enumerate(row):
                    loc = self.goal_world[i][j]
                    if(loc.char in block_dect):
                        g = '(block-at ' + block_dect[loc.char].char +' '+ loc.location +')'
                        goals.append(g)
            goal += '(and ' + ' '.join(goals) + '))'
            f.write(goal_comment + '\n')
            f.write(goal + '\n\n')
            
            
            #Problem Define Done
            done_comment = '; Problem Define Done'
            done = ')'
            f.write(done_comment + '\n')
            f.write(done + '\n')
            
        
    
    def get_adjacents(self, r, c):
        adjacent = ''
        world = self.world
        max_r = len(self.world)
        max_c = len(self.world[0])
        
        dr = -1
        dc = 0
        if((0 <= (r+dr) < max_r) and (0 <= (c+dc) < max_c)):
            adjacent += '(adjacent ' + str(world[r][c].location) +' '+ str(world[r+dr][c+dc].location) +' '+ str(WorldDirections.UP) +') '
        
        dr = 1
        dc = 0
        if((0 <= (r+dr) < max_r) and (0 <= (c+dc) < max_c)):
            adjacent += '(adjacent ' + str(world[r][c].location) +' '+ str(world[r+dr][c+dc].location) +' '+ str(WorldDirections.DOWN) +') '
        
        dr = 0
        dc = -1
        if((0 <= (r+dr) < max_r) and (0 <= (c+dc) < max_c)):
            adjacent += '(adjacent ' + str(world[r][c].location) +' '+ str(world[r+dr][c+dc].location) +' '+ str(WorldDirections.LEFT) +') '
        
        dr = 0
        dc = 1
        if((0 <= (r+dr) < max_r) and (0 <= (c+dc) < max_c)):
            adjacent += '(adjacent ' + str(world[r][c].location) +' '+ str(world[r+dr][c+dc].location) +' '+ str(WorldDirections.RIGHT) +') '
        
        adjacent += '\n'
        return adjacent
        
        
```
